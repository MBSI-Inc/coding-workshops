## 2.2) Slice

Now we are able to access individual characters in a string. What if we want to access a sequence of letters of a string like a word or a substring. We can do this by concatenating each individual character of this sequence. This, however, quickly becomes cumbersome if the substring is longer than a few characters.

Python provides a convenient method to access a substring: you simply need to specify a start and an end index. This method is also known as **slicing**. For example, to access individual words in `"MBSI Coding"`:

````python
s = "MBSI Coding"

# print "MBSI"
print(s[0] + s[1] + s[2] + s[3])  # you can concatenate individual characters
print(s[0:5])

# print "Coding"
print(s[3:8])
````

Note that the slice is always made before the end index. So a slice [a:b] will only include character at index b-1 but not index b

By default, slicing assumes you will slice from the start of the string to the end of the string. If you are slicing from either end, you can just be lazy and do this:

````python
s = "MBSI Coding"

# print "MBSI"
print(s[:4])     # no initial index provided

# print "Coding"
print(s[5:])     # no final index provided

# a slice from beginnning to end
print(s[:])
````

You can specify a third number which indicates the step size of your slice. For example, if you want every second element you can do this:

````python
example = "0123456789"
# slice the string by stepping through every second character
print(example[0:10:2])
print(example[::2])

# you can specify a range with the steps
print(example[2:8:2])
````

You can also make your step size negative to change the direction of slicing. For example:

````python
example = "0123456789"

# slice between index 8 to 4 from right to left:
print(example[8:4:-1])
````

## 3.2) Finding an item
`index` is a non-mutating method that returns the position of, again, the **first occurrence** of an item in a list, as seen in the following example, where the list contains multiple instances of `"Hib"`


````python
vaccinations = ["HepB", "HepB", "RV", "Hib", "RV", "Hib"]
print(vaccinations.index("Hib"))
````

An error occurs if the argument to index is not in the list:

````python
vaccinations = ["HepB", "HepB", "RV", "Hib", "RV", "Hib"]
print(vaccinations.index("HepC"))
````

To avoid bumping into an error everytime, we can check if the item we're searching for is in the sequence to begin with. We can do this using the `in` operator. Here is a function that searches for an item index in a sequence, and if the index is not found, it returns `None` instead of providing an error.

````python
def advanced_index(lst, item):
    if item in lst: 
        # this line checks if the item is in the list input
        return lst.index(item)
    else:
        return None

vaccinations = ["HepB", "HepB", "RV", "Hib", "RV", "Hib"]
print(advanced_index(vaccinations, "HepC"))
````

## 3.3) Sorting

One list operation that you will find many applications for is sorting the elements. There are two ways of doing this, which you will inevitably confuse at some point because of the names being so similar: `.sort()` and `sorted`. You may have guessed from the way they are written, one of them is a method and the other is a function. We'll start with the `.sort()` **methods**: 

`.sort()` mutates the list by puting all items in the list into ascending numeric order

````python
randlist1 = [4, 1, 3.0, 2, 5]
randlist1.sort()
print(randlist1)
````

````python
randlist2 = [4, 1, 3.0, 2, 5]
randlist2.reverse()
print(randlist2)
````

# 4) Tuples and mutability

### 4.1) Introduction to tuples

**Tuples** are like the stubborn brother of lists: a tuple looks like a list, but it is *immutable*. This means you can't change it once you have created it.

````python
my_tuple = (6, 2, 3, 4, 5)
my_tuple[0] = 2 # error, can't change any element of tuple
sorted(my_tuple) # returns a new sorted list, my_tuple is unchanged
````

Why would someone use tuples instead of lists? By being immutable, tuples have these advantages:

  1. Tuples are faster than lists (not relevant at our level)
  2. It makes your data stored in tuples safer, as they are not accidentally modified
  3. Tuples can be used as dictionary keys, but not lists (Dictionaries covered later)

## 4.2) Mutability vs immutability

If the object is *mutable*, it can be modified (mutated) after creation, if it is *immutable*, it can't be modified after creation. If you are using *mutable* objects, beware of the following:

**Example**

````python
list1 = [1, 2, 3, 4, 5]   
list2 = list1                     # is list 2 is a copy of list one?
list2.append(6)
print("This is list1:", list1)
print("This is list2:", list2)
````

Notice how altering `list2` alters `list1` as well? This is because both `list1` and `list2` are referring to the same thing (a *mutable* list). Use `list1.copy()` to prevent this from happening.

````python
list1 = [1, 2, 3, 4, 5]   
list2 = list1.copy()      # is list 2 is a copy of list one?
list2.append(6)
print("This is list1:", list1)
print("This is list2:", list2)
````

The case described above will happen when you use any mutable object:
  - lists, sets, dictionaries(adding a new key:value set)

This will not happen when using immutable objects:
  - tuples, bool, int, float, string

**Example**

In the following example we are using strings but feel free to use any of the data types used above. Notice how what happened above to lists don't happen to strings and immutables.

````python
str1 = "BMI"
str2 = str1
str2 += "f"
print("This is str1:", str1) # str1 is unchanged
print("This is str2:", str2) # str2 is changed
````

## 5.1) Dictionary

Now let's talk about a very useful data type: **dictionaries**. Dictionaries allow you to create mappings from **keys** to values. For example, you might use a dictionary to map patients to their corresponding conditions: 

````python
conditions = {'Jerry': 'Cystic Fibrosis',
    'Jack': 'Hypothermia',
    'Aaron': 'Gall Stones',
    'Lindsay': 'Jaundice',
    'Garry': 'COPD',
    'Amy': 'Coeliac'}

print(conditions["Jerry"])
````

The keys in the above dictionary are strings which represent the names of the patients. Each key maps to a value, which also happens to be a string (the patient's condition) in this case. In the general case, the keys and values can be of different **types**.

Dictionary values are constructed using the curly brace characters `{  }`. As a special case, you can construct an empty dictionary using an opening and closing brace with nothing in between:

`example_empty_dict = {}`

In the above code, an empty dictionary is created and assigned to the variable `example_empty_dict`. Currently this dictionary contains no mappings, so it is fairly uninteresting. As you shall soon see, it is possible to add new mappings to an existing dictionary, so it can be useful to construct empty dictionaries and later add values as necessary.

### 5.1.1) Dictionary indexing 

Like lists and strings, dictionaries are indexable. However, you index dictionaries by using a key value stored in the dictionary, it is different to indexing lists where you are using a position number. **Note that there isn't duplicate key values in dictionaries.**

````python
conditions = {'Jerry': 'Cystic Fibrosis',
    'Jack': 'Hypothermia',
    'Aaron': 'Gall Stones',
    'Lindsay': 'Jaundice',
    'Garry': 'COPD',
    'Amy': 'Coeliac'}

print(conditions["Jerry"])
print(conditions["Aaron"])
print(conditions["Carl"]) # error, Carl is not in the dictionary, he's healthy
print(conditions[2]) # error, the dictionary doesn't use ordering
````

### 5.1.2) Updating Dictionaries

You can change the value which is associated with a key or add a new key and value pair to the dictionary using the assignment (`=`) operator:

````python
conditions = {'Jerry': 'Cystic Fibrosis',
    'Jack': 'Hypothermia',
    'Aaron': 'Gall Stones',
    'Lindsay': 'Jaundice',
    'Garry': 'COPD',
    'Amy': 'Coeliac'} # Actually Amy has Ulcerative Colitis

# Let's change Amy's condition
conditions["Amy"] = "Ulcerative Colitis"

# Let's also add a new patient who has Hepatitis C
conditions["Irene"] = "Hepatitis C"

print(conditions["Amy"])
print(conditions["Irene"])
````

### 5.1.3) Membership and Looking at the entire dictionary

````python
conditions = {'Jerry': 'Cystic Fibrosis',
    'Jack': 'Hypothermia',
    'Aaron': 'Gall Stones',
    'Lindsay': 'Jaundice',
    'Garry': 'COPD',
    'Amy': 'Coeliac'}

# Check if Amy is a patient
print("Amy" in conditions)  # Try some other names

# Printing the dictionary
print(conditions.keys())    # All the keys
print(conditions.values())  # All the values
print(conditions.items())   # All the key:value pairs
````

# 6) `for` loops



## 6.1) Introduction to `for` loops

A typical `for` loop is of the following form:

```
for <item> in <sequence>:
    <instruction>
```

Basically, there is a *predetermined* sequence, e.g. a list. For each item in the sequence, the `for` loop runs once. Therefore, the length of the sequence specifies the total number of "iterations" through the loop. Note that `<item>` actually becomes a variable (also known as an "iterator") that can be used within the loop.

Confused? Let's see an example.

**Example**

````python
listOfNames = ["Alice", "Bob", "Charlie", "Dorothy"]

for name in listOfNames:
    print(f"Hi {name}!")
````

What happened above?

For the first "iteration" of the loop, the value `"Alice"` is assigned to the variable `name`. The print function within the loop executes. Then, the loop repeats, but this time `"Bob"` is assigned to `name`. The print function executes again in this second iteration.

This way, the loop runs four iterations; during each iteration, the items within `listOfNames` are assigned to `name` one-by-one.

Note also that the commands `continue` and `break` work for `for` loops as well. In the case of `continue`, when the loop resets, the `<item>` variable moves on to the next item in the sequence.

**Example**

````python
listOfNames = ["Alice", "Bob", "Charlie", "Dorothy"]

for name in listOfNames:
    if name == "Bob":
        print("Not you Bob")
        continue    # skips saying Hi to Bob and moves on to Charlie
    print(f"Hi {name}!")
````

## 6.2) `for` loops and dictionaries 
`for` loops can also be used to iterate through dictionaries, even though they are not sequences as they are unordered. This is done by looping over a dictionary's *keys*, not its values or key:value pairs. You can retrieve the values stored in the dictionary via the keys.

**Example**

````python
capital_cities = {"Australia":"Canberra", "New Zealand":"Wellington", "China":"Beijing", "USA":"Washington DC", "Russia":"Moscow"}
for country in capital_cities: # country variable iterates over keys in capital_cities
    print(f"{country}: {capital_cities[country]}") # Prints key and value pairs
````

Alternatively, we can produce the same output as above by explicitly telling the `for` loop to iterate through the key:value pairs. We can use the `items()` method (introduced in Workshop 5) to return the key:value pairs in the dictionary as `(key, value)` tuples in a list. We can then loop over this list, making sure the iterator variable is now a tuple as well. The advantage of this method is that it's easier to reference values directly instead of using keys.

## 6.3) Indexing using `range()`
The `range()` function generates a sequence of integers. By default, the sequence starts from `0`, increments by `1` and stops *before* a specified integer `n` (i.e. the sequence goes up to `n-1`). The general form when using the `range()` function is `range(n)`. A more specific example is demonstrated below:

**Example**

````python
x = range(5) # Generates a sequence from 0 (default) to 4, incrementing by 1 (default)
for i in x:
    print(i)
````


**Example**

````python
x = range(3, 9) # Generates a sequence from 3 to 8, incrementing by 1 (default)
print(x)
for i in x:
    print(i)
````


**Example**

````python
statement = "I don't want to go to my Zoom lecture"
repetitions = range(10) # Generates a sequence from 0 to 9, incrementing by 1
for repeat in repetitions: 
    print(statement) # Prints the statement 10 times
````

Another reason why `range()` is useful when combined with `for` loops is *indexing*. If you loop through a sequence directly, the iterator variable will only store the values. In some cases however, we not only want to access the values in a sequence but also their indexed positions. `range()` solves this problem. 

Let's illustrate this with an example where we find the indices and characters in the string `"COVID-19"`. 

First, we find the number of characters in the string using the `len()` function. 

```
len("COVID-19")
```

Next, we pass the length of the string into the the `range()` function to produce a sequence of string indices rather than characters.

```
range(len("COVID-19"))
```

Finally, we include this sequence in a `for` loop with `index` as the iterator variable. Now we can iterate through the string indices and print out both the index and the character. The character can be accessed using `str[index]`.

**Example**

````python
str = "COVID-19"

for index in range(len(str)):
    print(f"{index}: {str[index]}") # Printing index and character in string
````

# 3) Nested loops
Similar to how we used nested conditional statements in Workshop 3, we can create a *nested loop* by putting a loop inside a loop.

The "inner loop" is executed to completion once for every iteration of the "outer loop". You can include as many levels of nesting as you want, as long as you remember to indent properly.

An example of a nested `for` loop is shown below, however, you can also have nested `while` loops or `for` loops within `while` loops and vice versa.

**Example**

````python
colours = ["red", "green", "blue"]
shapes = ["triangle", "square", "circle"]

for colour in colours: # Outer loop iterating through colours
    print(f'{colour} shapes:') # Print for each iteration of the outer loop
    for shape in shapes: # Nested inner loop iterating through shapes
        print(colour, shape) # Print combination of colour and shape for each iteration of the inner loop
    print('---------------') # Print for each iteration of the outer loop

````
